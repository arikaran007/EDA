```python
df1.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = df1.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x="Aggregate rating",y="Rating count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating count",hue='Rating color',data=ratings,palette = ['blue','red','orange','yellow','green','pink'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette = ['blue','red','orange','yellow','green','pink'])
```
